# Dynamic Programming
: 큰 문제를 작게 나누고, 같은 문제라면 한 번씩만 풀어 문제를 효율적으로 해결하는 알고리즘 기법

* 연산을 진행 때 연산 속도와 메모리 공간을 최대한으로 활용할 수 있는 효율적인 알고리즘을 작성해야한다.
* 2가지 방식: 
  * Top-Down(상향식): 큰 문제를 해결하기 위해 작은 문제를 호출한다
  * Bottom-Up(하향식): 단순히 반복문을 이용하여 소스코드를 작성하는 경우 작은 문제부터 차근차근 답을 도출한다.
* e.g) 피보나치 수열
* 다음 조건을 만족할 때 사용
  * 큰 문제를 작은 문제로 나눌 수 있다.
  * 작은 문제에서 구한 정답은 그것을 포함하는 큰 문제에서도 동일하다.

* 핵심 기법(DP table): Memoization[=Caching](이미 풀었던 문제들은 더 이상 연산할 필요가 없다고 가정하고 넘어가기)

* *setrecursionlimit()*: 재귀적인 표현을 쓸때 재귀 제한을 완화할 수 있는 python method
  

```python
# 피보나치 수열 소스코드(재귀적)
def fibo(x):
  if x==1 or x==2:
    return 1
  return fibo(x-1) + fibo(x-2)

print(fibo(99))
```

```python
# 피보나치 수열 소스코드(재귀적) with memoization
d = [0] * 100

def fibo(x):
  if x==1 or x==2:
    return 1
  if d[x] !=- 0:
    return d[x]

  d[x] = fibo(x-1) + fibo(x-2)
  return d[x]

print(fibo(99))
```

```python
# Bottom up
d = [0]*30001
x = int(input())

for i in range(2,x+1):
  d[i] = d[i-1] + 1
  if i%2 ==0:
    d[i] = min(d[i],d[i//2]+1)
  if i%3 == 0:
    d[i] = min(d[i],d[i//3]+1)
  if i%4 == 0:
    d[i] = min(d[i],d[i//5]+1)
print(d[x])
```
